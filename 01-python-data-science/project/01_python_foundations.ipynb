```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "traffic_type": ["Normal", "Normal", "Attack", "Normal", "Attack"],
    "packet_count": [120, 98, 450, 110, 380],
    "duration": [30, 25, 120, 28, 100]
}

df = pd.DataFrame(data)
df
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Select a single column
df["packet_count"]
```

```python
# Filter rows
attack_traffic = df[df["traffic_type"] == "Attack"]
attack_traffic
```

```python
df.groupby("traffic_type")["packet_count"].mean()
```

```python
df.groupby("traffic_type")["packet_count"].mean().plot(kind="bar")
plt.title("Average Packet Count by Traffic Type")
plt.ylabel("Packet Count")
plt.show()
```

## Notes

- DataFrames encourage column-oriented thinking.
- Groupby operations are powerful for comparisons.
- Simple visualizations can already highlight differences.
- This workflow maps directly to CICIDS-style analysis.
